**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
###Creating Price Ratio Function

def create_price_ratio(d):
    y_1998 = float(d['Average Price'][d['Year']==1998])
    y_2018 = float(d['Average Price'][d['Year']==2018])
    price_ratio = [y_2018/y_1998]
    return price_ratio


#Testing function on Burough Barnet
create_price_ratio(dfg[dfg['Burough']=='Barnet'])
```

```python
#Will want to save this ratio across all the buroughs. Creating an empty dictionary to store that information
final = {}

for x in dfg['Burough'].unique():
    burough = dfg[dfg['Burough'] == x]
    final[x] = create_price_ratio(burough)
    
print(final)
```

```python
#Assigning to a new data frame price_ratios
price_ratios = pd.DataFrame(final)

#Calling head on new data frame
price_ratios.head(5)
```

```python
#Need to transpose this data frame
transposed_price_ratios = price_ratios.transpose()

#Calling head on new transposed data frame
transposed_price_ratios.head(5)
```

```python
#Resetting Index on transposed_price_ratios
transposed_price_ratios = transposed_price_ratios.reset_index()
#Printing Head
transposed_price_ratios.head(5)
```

```python
#Adjusting names of columns
transposed_price_ratios.rename(columns = {'index':'Burough', 0:'2018 Price Ratio'}, inplace=True)

#Checking Head to see proper allocation
transposed_price_ratios.head(5)

```

```python
#Sorting rank of 2018 and then graphing
top = transposed_price_ratios.sort_values(by = '2018 Price Ratio', ascending = False)

print(top)
```

```python
#Graphical Output

price_ratio_graph = top[['Burough', '2018 Price Ratio']]. plot(kind = 'bar', x = 'Burough', y = '2018 Price Ratio')

price_ratio_graph.set_xticklabels(top.Burough)

```
