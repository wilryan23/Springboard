1. Collect data from the Franfurt Stock Exchange, for the ticker ADS, for the whole year 2017 (keep in mind that the date format is YYYY-MM-DD)

```python
#Completing task #1, however, in order to pull in AFX data we need to pay for access to the database. 
#Thus, will be investigating Adidas AG (i.e. ADS) as it is part of the free data set

#Loading in URL & API Key
url = 'https://www.quandl.com/api/v3/datasets/XFRA/ADS/data.json?&start_date=2017-01-01&end_date=2017-12-31&api_key='+API_KEY

r = requests.get(url)


```

```python

#Assigning to json_file_url
ADS_data = r.json()

#Printing contents of Json File
print(ADS_data)
```

2. Convert the returned JSON object into a Python dictionary.

```python
#Completing task #2. Converting JSON object into Python dictionary
ADS_data_clean = dict(ADS_data)

#Confirming conversion into Python dictionary
print(type(ADS_data_clean))
```

3. Calculate what the highest and lowest opening prices were for the stock in this period.

```python
#In order to calculate the highest and lowest opening prices throughout the year, will need to look at what key/value pairs are useful

#checking the keys of json_file_clean
ADS_data_clean.keys()

#Will need to take this one step further as it is nested
ADS_data_clean['dataset_data'].keys()
```

```python
#Judging by the initial print out of ADS_data, know that keys of interest are column_names and data
#Let's take a look at them separately from the other keys
ADS_data_clean['dataset_data']['column_names'],ADS_data_clean['dataset_data']['data']
```

```python
#Based off the output above we know that the Open values are the second item in the the list 'data'
#Saving 'data' to ADS_final_data
ADS_final_data = ADS_data_clean['dataset_data']['data']

#printing type of ADS_final_data
print(type(ADS_final_data))
```

```python
#Now have a list of data to work with and know that our open price is in the second[1] row
opening_price = [x[1] for x in ADS_final_data if x[1] !=None]
print('The highest ADS price in 2017 was', max(opening_price))
print('The lowest ADS price in 2017 was', min(opening_price))
```

4. What was the largest change in any one day (based on High and Low price)?

```python
#Know that our high price is in the third[2] row and our low price is in the fourth[3] row
high_change = [x[2] for x in ADS_final_data if x[2] !=None]

low_change = [x[3] for x in ADS_final_data if x[3] !=None]

largest_change = [x1 - x2 for (x1,x2) in zip(high_change, low_change)]

print('The largest change in any one day (based on High and Low price) was',max(largest_change),'.')
```

5. What was the largest change between any two days (based on Closing Price)?

```python
#Know that our closing price is in the the fifth[4] row
closing_price = [x[4] for x in ADS_final_data if x[4] != None]
closing_price_prior = [x[4] for x in ADS_final_data if x[4] != None][1:]

largest_closing_price_change = [x1 - x2 for (x1,x2) in zip (closing_price,closing_price_prior)]

print('The largest change between any two days (based on Closing Price) was', max(largest_closing_price_change), '.')
```

6. What was the average daily trading volume during this year?

```python
#Know that volume is in the sixth[5] row
volume = [x[5] for x in ADS_final_data if x[5] !=None]

avg_volume = sum(volume)/len(volume)

print('The average daily trading volume during the year was',avg_volume,'.')
```
